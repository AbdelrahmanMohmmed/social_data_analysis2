```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
url = "https://www.youtube.com/watch?v=kNuwCiuDz4E"

driver.get(url)
time.sleep(5)
comments=[]
while len(comments)<100:
    # scroll down to load comments
    for _ in range(5):
        driver.execute_script("window.scrollBy(0, 1200);")
        time.sleep(3)

    # find all comments
    comments.append(driver.find_elements(By.XPATH, "//ytd-comment-view-model"))

print("Total comments:", len(comments))

for c in comments:
    try:
        author = c.find_element(By.XPATH, './/a[@id="author-text"]/span').text
        text = c.find_element(By.XPATH, './/yt-attributed-string[@id="content-text"]').text
        time_posted = c.find_element(By.XPATH, './/span[@id="published-time-text"]/a').text
        likes = c.find_element(By.XPATH, './/span[@id="vote-count-middle"]').text
    except:
        continue

    print("-------------")
    print("Author:", author)
    print("Time:", time_posted)
    print("Likes:", likes)
    print("Comment:", text)

```
